```python
# solve dy/dt = 10 - 2y

import numpy as np
import matplotlib.pyplot as plt

# define slope
def y_prime(y):
    return 10 - 2*y

# specify number of points (grid)
N = 50
t = np.linspace(-2, 2, N)
dt = 1 / N      # time interval

y = np.zeros(N)
y[0] = 6        # initial condition
#-------------------------------------------------------
# apply the Euler algorithm
for i in range(1, N):
    y[i] = y[i - 1] + y_prime(y[i - 1]) * dt
#-------------------------------------------------------
print("Euler's method 50 steps y(10) =", y[10])
plt.xlabel('t')
plt.ylabel('y')
plt.plot(t, y,'bo', label = 'Euler\'s Method')

#-------------------------------------------------------
#apply Heun's method
for i in range(1, N):
    yend = y[i-1] + y_prime(y[i-1])*dt #use Euler to find this point    
    y[i] = y[i-1] + (y_prime(y[i-1])+ y_prime(yend))/2 * dt
#-------------------------------------------------------
print("Heun's method 50 steps y(10) = ", y[10])
plt.plot(t,y, 'c*', label = "Heun\'s Method")    
#plot ze exact solution
plt.legend()
plt.show()
```

Euler and Heun's method with 50 steps and t from -2 to 2

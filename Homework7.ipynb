```python
#1.satellite with boost
Web VPython 3.2

#define constants
G = 6.67E-11    #gravitational constant
AU = 1.5E11     #astronomical unit in meters
YEAR = 365.25*24*60*60      #seconds in a year

#create a scene to work in
scene = canvas(title = 'Kepler\'s First Law', background = color.white)

#create objects for the earth and the satellite
#sun = sphere(pos = vec(0,0,0), radius = 1E10, mass = 2E30, color = color.yellow, texture = textures.stucco)
earth = sphere(pos = vec(0,0,0), radius = 6.4E6, mass = 6E24, texture = textures.earth)
sat = sphere(pos = vec(1.1*earth.radius,0,0), mass = 1E4, color = color.cyan, radius =1E3)

#give the satellite a velocity
sat.vel = vec(0,1.1*sqrt(G*earth.mass/sat.pos.x),0)
sat.trail = curve(pos=sat.pos, color = sat.color)

#time step
h = 1E2

rmax = 1*earth.radius
rmin = 10*earth.radius 
#define key press ---------------------------------------------------------
def keyInput(evt):
    z = evt.key
    if 'up' in z:
        sat.vel = 1.01*sat.vel#speed up
    if 'down' in z:
        sat.vel = 0.99*sat.vel#slow down
    if 'left' in z:
        sat.vel.z = sat.vel.z + mag(sat.vel)/10 #move satilite plane
    if 'right' in z:
        sat.vel.z = sat.vel.z - mag(sat.vel)/10 #move satilite plane

scene.bind('keydown',keyInput)

#define acceleration of satellite-------------------------------------------
def acc(satpos):

    r = mag(sat.pos - earth.pos)

    return -G*earth.mass*(satpos - earth.pos)/r**3

#---------------------------------------------------------------------------

counter = 0

#def k1_vel(t_i, y_i)
#def k2_vel(t_i, h, y_i, k1vel)
#def k3_vel(t_i, h, y_i, k2vel)
#def k4_vel(t_i, h, y_i, k3vel)    

#animate----------------------------------------------------------------------

while True:

    r = mag(sat.pos - earth.pos)

    if r > rmax:

        rmax = r

    if r < rmin:

        rmin = r

    a = (rmin + rmax)/2   #calculate the semimajor axis

    P = sqrt(4*pi*pi*a**3/(G*earth.mass))/60     #orbital period in minutes

    E = 0.5*sat.mass*mag2(sat.vel) - G*earth.mass*sat.mass/r    #energy
    
    L = sat.mass*cross(sat.pos,sat.vel)    #angular momentum

    ecc = sqrt(1 + 2*E*mag2(L)/((G*earth.mass)**2*sat.mass**3))

   

    sat.vel = sat.vel + acc(sat.pos)*h

    sat.pos = sat.pos + sat.vel*h
    
    #sat.vel = sat.vel*(1/6)*(k1_vel + (2*(k2_vel)) + (2*(k3_vel)) + k4_vel)*h
    #sat.pos = sat.pos*(1/6)*(k1_pos + (2*(k2_pos)) + (2*(k3_pos)) + k4_pos)*h

    sat.trail.append(pos=sat.pos, color=sat.color)
    
    if counter >=1000:
        print('Mag of Ang Mom = {0:8.3E}, Energy = {1:8.3E}, Ecc = {2:8.3E}, Period = {3:8.3E}'.format(mag(L), E, ecc, P))
        counter = 0
    
    counter = counter + 1
    rate(100)
```
